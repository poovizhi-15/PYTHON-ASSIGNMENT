**Strings (Concatenation, Slicing and Other methods) :**

1. String Concatenation:

```python
string1="HELLO"
```

```python
name=input("Enter your Name:")
```

```python
result=string1+name
print(result)
```

```python
string3=", Welcome to Python Programming"
```

```python
final_result=result+string3
print(final_result)
```

2.String Slicing and Indexing:

```python
text="HELLOpoo, Welcome to Python Programming"
print("First character:",text[0])
print("Last character:",text[-1])
print("First 5 characters:",text[:5])
print("Last 11 characters:",text[:-11])
print("Reversed string:",text[::-1])
print("Extracted word:",text[text.index("Python"):text.index("Python")+6])


```

3. String Methods

```python
strM="Python beginner tutorial"
print("Uppercase:",strM.upper())
print("Lowercase:",strM.lower())
print("Capitalized:",strM.capitalize())
print("Count of 't':",strM.count('t'))
print("Replaced string:", strM.replace("Python", "Machine Learning"))
```

Tuples (Creation, Modification and Access)

```python
t1=(10,20,30)
t2=(40,50,60)
t_combine=t1+t2
print("Combined Tuple:", t_combine)
print("Repeated Tuple:", t_combine * 3)
print("First three elements:", t_combine[:3])
print("Last three elements:", t_combine[-3:])
```
